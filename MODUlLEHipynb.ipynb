Differential equation - a relationship between a function, f(x), its independent variable, x, and any number of its derivatives.
Ordinary differential equation (ODE) is a differential equation is in one dimension.

$F(x, f(x), \frac{df(x)}{dx}, ..., \frac{d^{n-1}f(x)}{dx^{n-1}}) = \frac{d^nf(x)}{dx^n}$


##Reduction of Order

In order to solve higher order differential equations, we need to *reduce the order* of the differential equation to first order.

The ODEs of interest can be written as the following

$f^n(t) = F(t, f(t), f^1(t), ..., f^{n-1}(t))$

##The Euler Method
Let $\frac{dS(t)}{dt} = F(t, S(t))$ be explicitly defined first order ODE and let t be a numerical grid of the interval [$t_0, t_f$] with the spacing h.

The linear approximation of S(t) around $t_j$ at $t_{j+1}$ is

$S(t_{j+1}) = S(t_j) + (t_{j+1}-t_j)\frac{dS(t_j)}{dt}$
