##Interpolation Problem Statement
The interpolation function is an estimation function ^y(x) such that ^y$(x_i) = y_i$ for every data value $x_i$ and dependent data value $y_i$.

##Linear Interpolation
The estimated point is assumed to lie on the line that connects the nearest point from left and right.

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")
```

```python
x = [0, 1, 2]
y = [1, 3, 2]

f = interp1d(x, y)
y_hat = f(1.5)
print(y_hat)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(1.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

##Cubic Spline Interpolation
Similar to the linear interpolation except the funtion is a set of piecewise cubic functions.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")
```

```python
x = [0, 1, 2]
y = [1, 3, 2]

# use bc_type = 'natural' adds the constraints as we described above
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
