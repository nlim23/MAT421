##Interpolation Problem Statement

Interpolation is the technique that uses a function that estimates measurements based on the data points given.



##Linear Interpolation

It is assumed each given point are connect by a line from left to right.

Below is a example of using linear interpolation to find y for x  = 0.5 in the set of data x = [0, 1, 2], y = [1, 3, 2].


```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')

x = [0, 1, 2]
y = [1, 3, 2]

f = interp1d(x, y)
y_hat = f(0.5)
print(y_hat)
```

Below is the graph of the linear interpolation example.

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(0.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

##Cubic Spline Interpolation

Cubic spline interpolation is similar to linear interpolation, but the given points are connected by cubic functions.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')

x = [0, 1, 2]
y = [1, 3, 2]

# use bc_type = 'natural' adds the constraints as we described above
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
