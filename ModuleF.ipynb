##Numerical Differentiation Problem Statement

Numerical grid: Over some interval, an evenly spaced set of points over the domain of the independent variable of a function.

Spacing/Step size: The distance between adjacent points on the grid.

A function f(x) can be shown over an interval by computing the value onto a grid in Python. The function can be continuous, but the representation is discrete and this can be useful for engineering and in science practices.  

##Finite Difference Approximating Derivatives
The derivative f'(x) of a function f(x) at the point x = a is defined as:

f'(x) = lim$_x$$_→$$_a$$\frac{f(x)-f(a)}{x-a}$

The forward difference to estimate the slope of the function at x$_j$ using the line that connects ($x_j$, f($x_j$)) and ($x_j$$_+$$_1$, f($x_j$$_+$$_1$)):

f'($x_j$) = $\frac{f(x_j+_1)-f(x_j)}{(x_j+_1)-x_j}$

The backward difference to estimate the sloe of the function at $x_j$ using the line that connects ($x_j$$_-$$_1$, f($x_j$$_-$$_1$)) and ($x_j$, f($x_j$)):

f'($x_j$) = $\frac{f(x_j)-f(x_j-_1)}{x_j-(x_j-_1)}$

The central difference to estimate the slope of the function at $x_j$ using the line that connects ($x_j$$_-$$_1$, f($x_j$$_-$$_1$)) and ($x_j$$_+$$_1$, f($x_j$$_+$$_1$)):


f'($x_j$) = $\frac{f(x_j+_1)-f(x_j-_1)}{(x_j+_1)-(x_j-_1)}$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
%matplotlib inline
```

```python
# step size
h = 0.1
# define grid
x = np.arange(0, 2*np.pi, h)
# compute function
y = np.cos(x)

# compute vector of forward differences
forward_diff = np.diff(y)/h
# compute corresponding grid
x_diff = x[:-1:]
# compute exact solution
exact_solution = -np.sin(x_diff)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()

# Compute max error between
# numerical derivative and exact solution
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

##Approximating of Higher Order Derivatives

It is also possible to use the Taylor series to approximate higher order derivatives like f' '(x), f' ' '(x) and so on.

f(x$_j$$_-$$_1$) + f(x$_j$$_+$$_1$) = $\frac{f(x_j+_1) - 2f(x_j) + f(x_j-_1)}{h^2}$


```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
%matplotlib inline

x = np.arange(0, 2*np.pi, 0.01)
# compute function
omega = 100
epsilon = 0.01

y = np.cos(x)
y_noise = y + epsilon*np.sin(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'cos(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```
