### Base-N and Binary

Binary Numbers only have 0 and 1 digits.

```python
print(bin(100))

100 == 64 + 32 + 4
100 == 1*2**6 + 1*2**5 + 0*2**4 + 0*2**3 + 1*2**2 + 0*2**1 + 0*2**0
```

Adding binary numbers

```python
print(bin(17))
17 == 16 + 1
17 == 1*2**4 + 0*2**3 + 0*2**2 + 0*2**1 + 1*2**0

x = 0b1100100
y = 0b10001

print('0b1100100 + 0b10001 =', bin(117))
117 == x + y
```

Multiplying binary numbers

```python
x * y
print('0b1100100 * 0b10001 =', bin(1700))
1700 == x * y
```

### Round-off Errors
The difference between the true value and the number computed.

```python
5 - 4.945 == 0.055
```

The previous statement is considered false due to roundoff error. Below is the number computed from python.

```python
5 - 4.945
```

In order to show the true value, the rounding function needs to be used.

```python
round(5 - 4.945, 5) == round(0.055, 5)
```
